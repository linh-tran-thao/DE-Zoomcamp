```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql://postgres:postgres@localhost:5433/ny_taxi')
```

```python
df.head(n=0).to_sql(
    name='ny_taxi_data',
    con=engine,
    if_exists='replace'
)
```

```python
df.to_sql(
    name='ny_taxi_data',
    con=engine,
    if_exists='append'
)

print(f"Inserted data: {len(df)} rows")
```
